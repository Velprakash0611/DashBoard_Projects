# Undersampling

build a sample dataset containing similar distribution of legit and fraud transactions

No.of fraud transaction -->  271

```python
legit_sample = legit.sample(n=271)
```

# COncatenate the two datasets (fraud and legit_sample datasets)

```python
new_data = pd.concat([legit_sample, fraud], axis =0)

# axis =0 is used to concate the datasets in rowbase
# axis = 1 is used to concate column base
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

# Splitting the data into features and targets

```python
x=new_data.drop(columns='Class', axis=1)
y=new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

# Split the data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size =0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

#Model Training

```python
model = LogisticRegression()
```

```python
# training the model

model.fit(x_train,y_train)
```

#evaluation of model

Accuracy score

```python
# accuracy score on training data

x_train_prediction = model.predict(x_train)
training_data_accuracy= accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on training data == ',training_data_accuracy)
```

```python
# accuracy on test data

x_test_prediction = model.predict(x_test)
test_data_accuracy= accuracy_score(x_test_prediction, y_test)
print('Accuracy on test data = ', test_data_accuracy)
```
